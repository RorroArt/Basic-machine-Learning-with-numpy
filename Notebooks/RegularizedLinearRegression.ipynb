```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
import random as r
```

```python
data = load_diabetes()
```

```python
xt, yt = data.data[:222,:], (data.target[:222]/data.target.argmax())
xcv, ycv =  data.data[222:332,:], (data.target[222:332]/data.target.argmax())
xtest, ytest = data.data[332:442,:], (data.target[332:442]/data.target.argmax())
```

```python
def loss(y, yR):
    return np.sum((y - yR)**2)*(1/(2*(len(predict))))
```

```python
lossArrCV = []
lossArrTrain = []
numExamples = []
for i in range(37,223,37):
    x = np.append(np.ones((i,1)),xt[:i,:],axis=1)
    y = yt[:i]
    lr = 0.003
    lamb = 1
    b = np.zeros(11)
    predict = x@b
    for j in range(20000):
        regulGD = (lamb/y.size) * b
        regulGD[0] = 0.0
        grad = ((x.T @ (predict - y)) / y.size) + regulGD
        b -= (grad * lr)
        predict = x@b
        bRegul = b**2
        bRegul[0] = 0.0
        regulLoss = (lamb/(2*y.size)) * np.sum(bRegul)
        Tloss = np.sum((predict - y)**2)*(1/(2*(len(predict)))) + regulLoss
        if j % 500 == 0:
            plt.scatter(j,loss(predict, y))
    plt.show()
    lossArrTrain.append(loss(predict, y))
    predict = np.append(np.ones((xcv.shape[0],1)),xcv,axis=1)@b
    lossArrCV.append(loss(predict,ycv))
    numExamples.append(i)
```

```python
plt.plot(numExamples,lossArrTrain,label="Train")
print(lossArrTrain[-1])
plt.plot(numExamples,lossArrCV,label="cross validation")
print((lossArrCV[-1]))
plt.legend()
plt.show()
```

```python
lossArrCV = []
lossArrTrain = []
numExamples = []
for i in range(1,11):
    x = np.array([np.ones(i),xt[:i],xt[:i]**2]).T
    y = yt[:i]
    lr = 0.1
    lamb = 0
    b = np.zeros(3)
    predict = x@b
    for j in range(20000):
        regulGD = (lamb/y.size) * b
        regulGD[0] = 0.0
        grad = ((x.T @ (predict - y)) / y.size) + regulGD
        b -= (grad * lr)
        predict = x@b
        bRegul = b**2
        bRegul[0] = 0.0
        regulLoss = (lamb/(2*y.size)) * np.sum(bRegul)
        Tloss = np.sum((predict - y)**2)*(1/(2*(len(predict)))) + regulLoss
        if j % 500 == 0:
            plt.scatter(j,loss(predict, y))
    plt.show()
    plt.scatter(xt,yt,color='green',label='Train')
    plt.scatter(xcv,ycv,color='blue',label='Cross Validation')
    plt.scatter(xtest,ytest,color='red',label='Test')
    plt.plot(np.linspace(1,9)/9,(np.array([np.ones(50),np.linspace(1,9)/9,(np.linspace(1,9,)/9)**2]).T@b),color='black',label='Prediction')
    plt.legend()
    plt.show()
    lossArrTrain.append(loss(predict, y))
    predict = np.array([np.ones(xcv.size),xcv,xcv**2]).T@b
    lossArrCV.append(loss(predict,ycv))
    numExamples.append(i)
```

```python
plt.plot(numExamples,lossArrTrain,label="Train")
print(lossArrTrain[-1])
plt.plot(numExamples,lossArrCV,label="cross validation")
print((lossArrCV[-1]))
plt.legend()
plt.show()
```
